```python
from bs4 import BeautifulSoup

import urllib
```

```python
''' Quick scraper to grab potential list of ingredients '''

url = 'http://www.enchantedlearning.com/wordlist/food.shtml'

"""define opener"""
class MyOpener(urllib.FancyURLopener):
    version = 'Mozilla/5.0 (Windows; U; Windows NT 5.1; it; rv:1.8.1.11) Gecko/20071127 Firefox/2.0.0.11'
myopener = MyOpener()

page = myopener.open(url)
soup = BeautifulSoup(page, 'lxml')
```

```python
tables = soup.find_all('tr', {'align': 'center', 'valign':'top'})
```

```python
exclude_list = ['appetizer', 'appetite']
food_labels = []

for i in tables:
    text = i.get_text()
    text = text.split('\n')
    for j in text:
        text = str(j).lower()
        if text != '' and text != '\r' and text not in exclude_list:
            food_labels.append(text)


```
